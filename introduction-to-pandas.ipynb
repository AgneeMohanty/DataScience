## how to convert a column price to integer


```python
pepsi_sales.dtypes["Market Cap"]
```

```python
#this shows that instead of integers the values are objects
```

```python
pepsi_sales
```

```python
pepsi_sales["Market Cap"] = pepsi_sales["Market Cap"].astype(str).str.replace('[\$\,B]', '',regex=True)
#you can experiment with this later but what it basically does is it takes the values in market cap
# as strings or in form of strngs and then replaces the dollar,and b and in next step it stores the values in type float
pepsi_sales["Market Cap"] = pepsi_sales["Market Cap"].astype(float)
#this command takes the string in marketcap column and replaces the ($),(,),(.),with nothing and stores as type int.
```

```python
pepsi_sales
```

```python
pepsi_sales["Market Cap"].plot()
#now it is plottable
```
