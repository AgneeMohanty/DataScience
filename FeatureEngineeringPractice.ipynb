### Step2:Handling Missing Values

```python
#Lets first take a look which columns have how many null values and how many values of each datatype
df.info()
```

#### we can see that Alley,MasVnrType,FireplaceQu,PoolQC,Fence,MiscFeature are the columns having too many null values ,so its better we remove these features altogether.

```python
df.drop(columns=['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature'],inplace=True)
```

### We need to convert the categorical values into numerical ones using ENCODER
### Step3-Feature transformation

This part is being added after i went throught the imputation process and found that imputer can only handle numerical values,so for object datatype we must use encoders 


```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in df.columns:
    if df[column].dtype == 'object':#only convert columns of datatype object to categorical codes
        df[column]=df[column].fillna(df[column].mode())
        df[column] = label_encoder.fit_transform(df[column])

df.info()
```

#### Now lets take care of other missing values,for this we will use two strategies
* fill object columns with mode 
* fill selected int64 and float64 columns with mean
* fill selected int64 and float64 columns with mode

We are going to take a decision here, we notice that among the columns having numerical values , some have values varrying from 0 to 1,2,3 ,10 etc while some have bigger numbers, so for columns with number above 50 range we will use mean and for those below we shall use mode

For this we are going to use imputer tool

```python
from sklearn.impute import SimpleImputer

#Creating instances of SimpleImputer for mean and mode
mean_imputer = SimpleImputer(strategy='mean')
#mode_imputer =SimpleImputer(strategy='most_frequent')

#Setting up columns to use the imputer class on
cols_to_impute_mean=[col for col in df.columns if df[col].dtype in ['int64','float64'] and df[col].mean()>50]
#cols_to_impute_mode=[col for col in df.columns if df[col].dtype =='object' and df[col].mean()<50]
cols_to_impute_mean
#Fitting the imputer and transforming the Dataframe
df[cols_to_impute_mean] = mean_imputer.fit_transform(df[cols_to_impute_mean]) 
 
```
